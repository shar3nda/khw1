```python
import pandas as pd
sorts = pd.read_csv("/home/johann/CLionProjects/khw1/results.csv", sep=';')
sorts = sorts.groupby(['method', 'array_type', 'array_length'])['time_nanoseconds'].mean().reset_index()
sorts
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize=(10,8))
plt.ticklabel_format(style='plain')
for sort_method in sorts['method'].unique():
    cur = sorts.loc[(sorts['method'] == sort_method) & (sorts['array_type'] == 'RANDOM_0_5')]
    plt.plot(cur['array_length'], cur['time_nanoseconds'], label=sort_method)
    plt.title('Случ. 0-5')
    plt.xlabel('Длина массива')
    plt.ylabel('Время работы (нс)')
plt.legend()
```
